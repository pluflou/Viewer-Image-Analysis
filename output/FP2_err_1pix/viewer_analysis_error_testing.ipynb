```python
def findMean(xx, prof):
    avg_x = []
    norm = []

    avg_x = np.sum(np.multiply(xx, prof))
    norm = np.sum(prof)

    avg = avg_x/norm
    return avg

def findSTD(xx, prof, avg):

    mean_diff = 0
    weighted_diff = 0
    norm = 0


    mean_diff = (xx - avg)**2
    weighted_diff = np.abs(np.sum(prof*mean_diff))
    norm = np.abs(np.sum(prof))

    
    std = np.sqrt(weighted_diff/(norm*(1-1/len(prof))))
    
    return std
```
