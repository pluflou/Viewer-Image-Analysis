```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#first 2 were 5000 iter, last was 50000
stats = pd.read_csv("data_mean_stdev_noErr.txt", sep=' ', header= None)
stats_werr = pd.read_csv("data_mean_stdev_wErr.txt", sep=' ', header= None)
stats_3 = pd.read_csv("data_mean_stdev_3.txt", sep=' ', header= None)

stats_3.columns= ["name", "x_mean", "x_mean_std", "x_std_mean", "x_std_std", "y_mean", "y_mean_std", "y_std_mean", "y_std_std"]
stats.columns= ["name", "x_mean", "x_mean_std", "x_std_mean", "x_std_std", "y_mean", "y_mean_std", "y_std_mean", "y_std_std"]
stats_werr.columns= ["name", "x_mean", "x_mean_std", "x_std_mean", "x_std_std", "y_mean", "y_mean_std", "y_std_mean", "y_std_std"]
```

```python
x_pics =  np.arange(6,56)
```

P.S. 5 mm is 41 pixels

```python
plt.figure(figsize=(12,8))
plt.fill_between(x_pics, stats['x_mean'],stats['x_mean']+stats["x_mean_std"], alpha=0.6, edgecolor='b', facecolor='g', linewidth=0.6, antialiased=True)           
plt.fill_between(x_pics, stats['x_mean'],stats['x_mean']-stats["x_mean_std"], alpha=0.6, edgecolor='b', facecolor='g', linewidth=0.6, antialiased=True)

plt.xlabel("Image Number")
plt.ylabel("Pixels")
plt.title("Mean")

#plt.savefig("mean_images", dpi = 300)
```

```python
plt.figure(figsize=(12,8))
plt.fill_between(x_pics, stats['x_mean'],stats_werr['x_mean']+stats["x_std_std"], alpha=0.6, edgecolor='#1B2ACC', facecolor='#089FFF', linewidth=0.6, antialiased=True)           
plt.fill_between(x_pics, stats['x_mean'],stats_werr['x_mean']-stats["x_std_std"], alpha=0.6, edgecolor='#1B2ACC', facecolor='#089FFF', linewidth=0.6, antialiased=True)  

plt.xlabel("Image Number")
plt.ylabel("Pixels")
plt.title("STDEV")
```

```python
plt.figure(figsize=(12,8))

plt.plot(x_pics, stats['x_mean'], marker='.')
plt.plot(x_pics, stats_werr['x_mean'], color='orange', marker='.')
plt.plot(x_pics, stats_3['x_mean'], color='r', marker='.')


plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.plot(x_pics, stats['y_mean'], marker='.')
plt.plot(x_pics, stats_werr['y_mean'], color='orange', marker='.')
plt.plot(x_pics, stats_3['y_mean'], color='r', marker='.')

plt.ylim(0,250)

plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.plot(x_pics, stats['y_mean_std'], marker='.')
plt.plot(x_pics, stats_werr['y_mean_std'], color='orange', marker='.')
plt.plot(x_pics, stats_3['y_mean_std'], color='r', marker='.')

plt.ylim(0,30000)

plt.show()
```
