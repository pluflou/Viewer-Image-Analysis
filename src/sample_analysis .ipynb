## Raw data import 

```python
import numpy as np
import sys
import matplotlib as mpl
import matplotlib.pyplot as plt

import skimage.io as io 

# CCD images are in TIFF format
# importing the beam image 
raw_im = io.imread("D1542_2_22_9_22_Q3_-61_allother_nominal_043.tiff")
raw_im = io.imread("D1542_allnom_3_11_8pm_058.tiff")

# importing the background image
bg_im = io.imread("D1542_2_22_17_34_bg_053.tiff")
bg_im = io.imread("D1542_bg_3_11_8pm_057.tiff")

# look at images
plt.subplot(1, 3, 1)
plt.imshow(raw_im, cmap='cividis')

plt.subplot(1, 3, 2)
plt.imshow(bg_im, cmap='cividis')
```

## Background subtraction

The image is imported as an MxN array of numbers, where each array element corresponds to the numbers of counts in a pixel. The bagckground subtraction is done element by element.

```python
# first create an instance of an image
from im_reduction import Image

image = Image(raw_im)

# subtract background element by element
image.subtract_bg(bg_im);
```

```python
# plotting the subtracted image
plt.imshow(image.subtracted_data, cmap='cividis')
```

## Project image onto axes and integrate

To find the beam center and quantify the beam size, the problem can be simplified by projecting each array axis and integrating the values. By doing that, the beam distribution on each axis is treated separately as a 1D distribution study. 

The beam profile in each axis is found by taking the sum of the counts in each individual column along each axis.

```python
# calculate the x and y profiles
image.get_profile()

# get axis sizes
y = np.arange(image.y_size)
x = np.arange(image.x_size)

# assign variables
x_prof = image.profile_x
x_err = image.error_x
x_ax = x

y_prof = image.profile_y
y_err = image.error_y
y_ax = y
```

```python
# look at x-axis beam profile
plt.plot(x_prof)
```

## Find the center and width of the distribution

Several methods could be used to find the center. A widely used one is fitting a normal distribution (Gaussian) to the data and taking the mean of the distribution. However, due to noise and beam shapes that aren't consistently normally distributed, this approach is not robust.

A simpler method to achieve our goals is to take the median (location where the cumulative sum is 50% of the total sum of the distribution). This works well provided the background subtraction is well done. 

Similarly, the width of the distribution can defined as the $\pm 2\sigma$ (or whatever is more appropriate for the goal of the analysis/tracking). In this notebook, we take $\pm 2\sigma$ as the points where the cumulative sum reaches 15.8% and 84.2% of the total sum of the distribution.

```python
from im_reduction import find_median

x_mean, x_nsig, x_psig = find_median(x_prof)
y_mean, y_nsig, y_psig = find_median(y_prof) 

x_peak, x_idx= np.max(x_prof), list(x_prof).index(np.max(x_prof)) 
y_peak, y_idx= np.max(y_prof), list(y_prof).index(np.max(y_prof))
```
